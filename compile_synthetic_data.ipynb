```python
import pandas as pd

qas = pd.read_csv('./data/qa_dataset4_complete.csv')
qas.head()
```

```python
# print prompt
print(qas['prompt'][0])
```

```python
def clean_prompts(prompt):
    lines = prompt.split('<|im_start|>user')
    lines = lines[1:-1]
    lines = [l.strip() for l in lines]
    prompt =  ""
    for l in lines:
        prompt += "<|im_start|>user\n " + l.strip() + "\n"

    return prompt

# try

print(clean_prompts(qas['prompt'][0]))
```

```python
def clean_paragraph(paragraph):
    lines = paragraph.split('\n')
    lines = [l.strip() for l in lines]
    return ' '.join(lines)

# try
print(clean_paragraph(qas['paragraph'][0]))
```

```python
# create new dataframe with cleaned prompts and paragraphs

df = pd.DataFrame()
df['prompt'] = qas['prompt'].apply(clean_prompts)
df['paragraph'] = qas['paragraph'].apply(clean_paragraph)

# add file names
df['file_name'] = qas['file_name']

# add length of prompt
df['prompt_length'] = df['prompt'].apply(lambda x: len(x.split(' ')))

# add length of paragraph
df['paragraph_length'] = df['paragraph'].apply(lambda x: len(x.split(' ')))

# add difference in length
df['length_difference'] = df['paragraph_length'] - df['prompt_length']
```
